```python
import torchio as tio
```

```python
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [20, 40]
```

```python
rescale = tio.RescaleIntensity((0, 255), percentiles=(1, 99))
to_ras = tio.ToCanonical()
colin = rescale(tio.datasets.Colin27().t1)
fpg = rescale(to_ras(tio.datasets.FPG().t1))
```

```python
colin.spacing, fpg.spacing
```

```python
colin.shape, fpg.shape
```

```python
resize = tio.Resize(256)
colin256 = resize(colin)
fpg256 = resize(fpg)
```

```python
colin256.plot()
```

```python
fpg256.plot()
```

```python
colin256.set_data(colin256.data[..., 128:129, :].permute(0, 1, 3, 2))
fpg256.set_data(fpg256.data[..., 128:129, :].permute(0, 1, 3, 2))
```

```python
colin256.shape, fpg256.shape
```

```python
colin256.as_pil().rotate(180)
```

```python
fpg256.as_pil().rotate(180)
```

### Resample

```python
resample = tio.Resample(2)
colin2 = resample(colin)
fpg2 = resample(fpg)
```

```python
colin2.plot()
```

```python
fpg2.plot()
```

```python
colin2.spacing, fpg2.spacing
```

```python
colin2.shape, fpg2.shape
```

```python
colin2.set_data(colin2.data[..., 64:65, :].permute(0, 1, 3, 2))
fpg2.set_data(fpg2.data[..., 64:65, :].permute(0, 1, 3, 2))
```

```python
colin2.as_pil().rotate(180)
```

```python
fpg2.as_pil().rotate(180)
```
